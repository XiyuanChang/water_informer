```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "04193500_real_pre_tp.xlsx"
df = pd.read_excel(file_path)
start_date = pd.to_datetime("1982-01-01")
end_date = pd.to_datetime("2018-12-31")
date_range = pd.date_range(start=start_date, end=end_date, periods=len(df))[0:100]

y_scatter = df.iloc[:, 0][0:100]
y_curve = df.iloc[:, 1][0:100]

print(y_scatter.shape)
print(y_curve.shape)
plt.figure(figsize=(30, 6))

plt.plot(date_range, y_curve, label="Prediction", color='black', linestyle='-')

plt.scatter(date_range, y_scatter, label="Real data", color='r', s=10)

plt.title("FlowAveraged_TP_mg/L vs 00065 pred")
plt.xlabel("Date") 
plt.ylabel("Values")  
plt.xticks(rotation=45) 
plt.legend()
plt.grid(True)

plt.show()
```
